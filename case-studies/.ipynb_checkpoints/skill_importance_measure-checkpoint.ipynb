```python
import networkx as nx
import oplc.model
from oplc.core import job_recommendation
import matplotlib.pyplot as plt
import numpy as np
import pandas as pa
```

```python
%matplotlib widget
plt.ioff()
```

```python
model = oplc.model.get()
```

```python
def show_graph(experience_names):
    exp_from_name = {exp.name: exp_id
                     for exp_id, exp in model.experiences.items()}

    experience_ids = [
        exp_from_name[name]
        for name in experiences
    ]

    print("Selected Experiences:")
    for e in experience_ids:
        print(f"{e}: {model.experiences[e].name} ({model.experiences[e].exp_type})")

    jr = job_recommendation(
        model.experiences_skills,
        model.jobs_skills,
        experience_ids,
        return_graph=True,
    )

    nodes = pa.DataFrame({
        "x": pa.Series({n: x for n, [x, _] in jr.skill_graph.layout.items()}),
        "y": pa.Series({n: y for n, [_, y] in jr.skill_graph.layout.items()}),
        "centrality": pa.Series(jr.skill_graph.centrality),
    })
    
    fig, ax = plt.subplots(1, 1)

    for u, v in jr.skill_graph.graph.edges:
        if u != v:
            ax.plot(
                [nodes.loc[u, "x"], nodes.loc[v, "x"]],
                [nodes.loc[u, "y"], nodes.loc[v, "y"]],
                color="black",
                linewidth=1,
                zorder=1,
            )
    s = ax.scatter(nodes.x, nodes.y, c=nodes.centrality, s=100,
                zorder=2, cmap="viridis")
    fig.colorbar(s)
    display(fig.canvas)
```

```python
experiences = [
    "Maintenance électrique",
    "Montage et câblage électronique",
    "Installation et maintenance d'ascenseurs",
]
show_graph(experiences)
```

```python
experiences = [
    "Animateur / Animatrice 2D",
    "Lead graphiste - jeux vidéo",
    "UX - user experience designer",
    "UI - user interface designer",
]
show_graph(experiences)
```
